```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("twitter_train.csv",delimiter = ",")
```

```python
x_text = data["text"]
```

```python
set(data["airline_sentiment"])
```

```python
data.head()
```

```python
x_text = list(x_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer(max_df = 0.9, min_df = 0.001, ngram_range = (1,2))
x_train = count_vec.fit_transform(x_text)
x_train.todense()
```

```python
x_train.shape
```

```python
def change_to_numeric(sentiment):
    if sentiment == 'negative':
        return 0
    elif sentiment == 'neutral':
        return 1
    else:
        return 2
```

```python
y_train_labelled = list(data["airline_sentiment"])
y_train = []
for i in y_train_labelled:
    y_train.append(change_to_numeric(i))
y_train
```

```python
x_train = x_train.todense()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST = train_test_split(x_train,y_train,random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha = 0.1)
clf.fit(X_TRAIN,Y_TRAIN)
```

```python
clf.score(X_TEST,Y_TEST)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_TRAIN,Y_TRAIN)
```

```python
clf.score(X_TEST,Y_TEST)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_TRAIN,Y_TRAIN)
```

```python
clf.score(X_TEST,Y_TEST)
```

```python
def change_to_labelled(sentiment):
    if sentiment == 0:
        return 'negative'
    elif sentiment == 1:
        return 'neutral'
    else:
        return 'positive'
```
